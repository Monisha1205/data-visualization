# Univariate analysis

```python
import seaborn as sb
sb.distplot(dataset["hsc_p"],hist=True,kde=True,kde_kws={'color':'red'})
```

# Bivariate analysis : distribution plot

```python
sb.jointplot(x='hsc_p',y='salary',data=dataset)
```

```python
sb.jointplot(x='hsc_p',y='salary',data=dataset,kind='hex')
```

```python
sb.jointplot(x='hsc_p',y='salary',data=dataset,kind='kde')
```

```python
# pair plot
```

```python
sb.pairplot(dataset,hue="gender",diag_kind="kde",kind="scatter",palette='husl')
```

# for categorical plot

```python
sb.stripplot(x='hsc_b',y='hsc_p',data=dataset)
```

```python
sb.stripplot(x='hsc_b',y='hsc_p',data=dataset,jitter=True)
```

```python
sb.swarmplot(x='hsc_b',y='hsc_p',data=dataset)
```
